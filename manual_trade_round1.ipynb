```python
import numpy as np
```

```python
er = np.array([[1, 0.5, 1.45, 0.75],
               [1.95, 1, 3.1, 1.49],
               [0.67, 0.31, 1, 0.48],
               [1.34, 0.64, 1.98, 1]])
```

```python
choices = np.asarray([0, 1, 2, 3])
combs = np.array(np.meshgrid(choices, choices, choices, choices)).T.reshape(-1, 4)
print(combs)
```

```python
profits = []

for comb in combs:
    i, j, k, l = comb
    profits.append(er[3, i] * er[i, j] * er[j, k] * er[k, l] * er[l, 3])
```

```python
max(profits)
```
