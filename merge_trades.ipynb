```python
import pandas as pd

t1 = pd.read_csv("round1/trades_round_1_day_-2_nn.csv", delimiter=";").drop(columns="profit_and_loss")
t2 = pd.read_csv("round1/trades_round_1_day_-1_nn.csv", delimiter=";").drop(columns="profit_and_loss")
t2["timestamp"] += int(1e6)

bt1 = t1.drop(t1[t1["symbol"] == "PEARLS"].index)
bt2 = t2.drop(t2[t2["symbol"] == "PEARLS"].index)
pt1 = t1.drop(t1[t1["symbol"] == "BANANAS"].index)
pt2 = t2.drop(t2[t2["symbol"] == "BANANAS"].index)

bt = pd.concat([bt1, bt2], axis=0).reset_index(drop=True)
pt = pd.concat([pt1, pt2], axis=0).reset_index(drop=True)

bt.to_csv("round1/trd_bnn_train.csv", index=False, sep=";")
pt.to_csv("round1/trd_prl_train.csv", index=False, sep=";")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['axes.formatter.useoffset'] = False

fig, ax = plt.subplots(2, 1)
ax[0].plot(bt["timestamp"])
ax[1].plot(pt["price"])
```
