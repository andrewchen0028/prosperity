```python
import pandas as pd

p1 = pd.read_csv("round1/prices_round_1_day_-2.csv", delimiter=";")
p2 = pd.read_csv("round1/prices_round_1_day_-1.csv", delimiter=";")
p2["timestamp"] += int(1e6)

bp1 = p1.drop(p1[p1["product"] == "PEARLS"].index).drop(columns=["day"])
bp2 = p2.drop(p2[p2["product"] == "PEARLS"].index).drop(columns=["day"])
pp1 = p1.drop(p1[p1["product"] == "BANANAS"].index).drop(columns=["day"])
pp2 = p2.drop(p2[p2["product"] == "BANANAS"].index).drop(columns=["day"])


bp = pd.concat([bp1, bp2], axis=0).reset_index(drop=True)
pp = pd.concat([pp1, pp2], axis=0).reset_index(drop=True)

bp.to_csv("round1/round1_banana_prices_outofsample.csv", index=False, sep=";")
pp.to_csv("round1/round1_pearl_prices_outofsample.csv", index=False, sep=";")
```
