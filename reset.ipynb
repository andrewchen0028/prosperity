```python
import pandas as pd
```

```python
ob = pd.read_csv("round1/ob_bnn_train.csv", delimiter=";",
                 index_col="timestamp")
trd = pd.read_csv("round1/trd_bnn_train.csv", delimiter=";",
                  index_col=None)

trd
```

```python
directon = []

for row in trd.to_numpy():
    ts = int(row[0])

    if row[5] <= ob.loc[ts, 'bid_price_1']:
        directon.append('S')
    
    elif row[5] >= ob.loc[ts, 'ask_price_1']:
        directon.append('B')
    
    elif ob.loc[ts, 'bid_price_1'] < row[5] < ob.loc[ts, 'ask_price_1']:
        directon.append('M')
    else:
        directon.append('E')

trd['dir'] = directon
```
